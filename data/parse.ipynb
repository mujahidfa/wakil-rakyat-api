```python
import csv
import json
```

```python
# Parse all calon DUN
calon_dun = []

with open('./raw/keputusan-pru-14-dewan-undangan-negeri.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    line_count = 0
    for row in csv_reader:
        calon_dun.append(row)
        
calon_dun
```

```python
# Get calon who won the DUN kerusi
dun = [i for i in calon_dun if i['STATUS'] in ['MENANG','MENANG TANPA BERTANDING'] or i['NEGERI'] == 'SARAWAK']
dun
```

```python
# Rename dict keys in the DUN dict
new_dun = []
for d in dun:
    
    bahagian = d['BAHAGIAN PILIHAN RAYA'].split(" - ")
    
    nombor_kod = int(bahagian[0][3:5])
    kod = bahagian[0].replace(" ", "")
    kawasan = bahagian[1]
    
    new_dun.append({
        "nombor_kod": nombor_kod,
        "kod": kod,
        "kawasan": kawasan,
        "negeri": d["NEGERI"],
        "nama_wakil_rakyat": d['NAMA CALON'].strip(),
        "parti": d["PARTI"],
    })

new_dun
```

```python
assert len(new_dun) == 587
```

```python
parlimen_dun_raw = []

with open('./raw/senarai-bahagian-pilihanraya-dun-pru-14.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    line_count = 0
    for row in csv_reader:
        if row not in parlimen_dun_raw:
            parlimen_dun_raw.append(row)

parlimen_dun_raw
```

```python
parlimen_dun = {}

for p in parlimen_dun_raw:
    
    parlimen_id = int(p['KOD DAN NAMA PARLIMEN'][2:5])
    dun_id = int(p['KOD DAN NAMA DUN'][3:5])
    
    if parlimen_id not in parlimen_dun:
        parlimen_dun[parlimen_id] = []
        
    parlimen_dun[parlimen_id].append(dun_id)

parlimen_dun
```

```python
parlimen_raw = []

with open('./raw/keputusan-pru-14-parlimen.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    line_count = 0
    for row in csv_reader:
        parlimen_raw.append(row)

parlimen_raw = [i for i in parlimen_raw if i['STATUS'] in ["MENANG"]]

parlimen = []

for p in parlimen_raw:
    bahagian = p['BAHAGIAN PILIHAN RAYA'].split(" - ")
    nombor_kod = int(bahagian[0][3:6])
    kod = bahagian[0].replace(" ", "")
    kawasan = bahagian[1]
    
    parlimen.append({
        "nombor_kod": nombor_kod,
        "kod": kod,
        "kawasan": kawasan,
        "negeri": p["NEGERI"],
        "nama_wakil_rakyat": p['NAMA CALON'].strip(),
        "parti": p["PARTI"],
    })

parlimen
```

```python
for p in parlimen:
    kod_parlimen = p['nombor_kod']
    p['dun'] = []
    if kod_parlimen in [125,166] or kod_parlimen in range(114, 125):
        continue
    else:
        kod_dun = parlimen_dun[kod_parlimen]
        for k in kod_dun:
            curr_dun = [d for d in new_dun if d['nombor_kod'] == k and d['negeri'] == p['negeri']][0]
            curr_dun["kod_parlimen"] = kod_parlimen
            p['dun'].append(curr_dun)
 
parlimen = sorted(parlimen, key = lambda i: i['nombor_kod'])
parlimen
```

```python
with open("./clean/data.json", "w") as outfile:  
    json.dump(parlimen, outfile) 
```
