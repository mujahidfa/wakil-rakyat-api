```python
calon_dun = []

with open('../keputusan-pru-14-dewan-undangan-negeri.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    line_count = 0
    for row in csv_reader:
#         if line_count == 0:
#             print(f'Column names are {", ".join(row)}')
#             line_count += 1
        calon_dun.append(row)
#         print(row)
#         print(f'\t{row["NEGERI"]}, {row["BAHAGIAN PILIHAN RAYA"]}, {row["NAMA CALON"]}, {row["NAMA CALON"]}, {row["NAMA CALON"]}, {row["NAMA CALON"]}.')
        line_count += 1
    print(f'Processed {line_count} lines.')
```
